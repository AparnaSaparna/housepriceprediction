```python
## import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading housing csv file
data=pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
# drop the rows and columns
data.dropna()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
#Selecting features using the above listed methods for the regression problem of predicting the “MEDV” column
from sklearn.model_selection import train_test_split
X=data.drop(['MEDV'],axis=1)
y=data['MEDV']
```

```python
X
```

```python
y
```

```python
#split your dataset in supervised machine learning
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
train_data=X_train.join(y_train)
```

```python
from sklearn.linear_model import LinearRegression
X_train,y_train=train_data.drop(['MEDV'],axis=1),train_data['MEDV']
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
forest=RandomForestRegressor()
forest.fit(X_train,y_train)
```

```python
forest.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
forest=RandomForestRegressor()
param_grid={
    "n_estimators":[3,10,30],
    "max_features":[2,4,6,8]
}
grid_search=GridSearchCV(forest,param_grid,cv=5,scoring="neg_mean_squared_error",return_train_score=True)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
#retrieve the best model and then use the score function of your estimator to calculate the train and test accuracy
grid_search.best_estimator_.score(X_test,y_test)
```
